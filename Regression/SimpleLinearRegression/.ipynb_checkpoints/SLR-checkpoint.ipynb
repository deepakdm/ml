```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import Data set
dataset = pd.read_csv('Salary_data.csv')

X = dataset.iloc[:,:-1].values # [:,:-1]
y = dataset.iloc[:,1].values
```

```python
# Splitting data set into Training set & Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
# Simple Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predict
y_pred = regressor.predict(X_test)
y_value = regressor.predict(15)
```

```python
# Visusalisation of Test data
plt.scatter(X_test, y_test, color='r')
plt.plot(X_test, y_pred, color='b')
plt.title('Salary v/s Expectation')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```

```python
# Visusalisation of Train data
plt.scatter(X_train, regressor.predict(X_train), color='r')
plt.plot(X_train, regressor.predict(X_train), color='b')
plt.title('Salary v/s Expectation')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```
