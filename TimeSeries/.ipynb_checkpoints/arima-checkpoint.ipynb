```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
def adf_check(time_series):
    result = adfuller(time_series)
    print('Augmented Dickey Fuller Test:')
    labels=['ADF Test Statistics', 'p-value', '#Lags used', 'Number of observations used']
            
    for value, label in zip(result, labels):
        print(label + ':' + str(value))
    
    if(result[1] <= 0.05):
        print('Strong evidence against the null hypothesis, rejects null hypothesis. Data has no unit root and stationary')
    else:
        print('It is non stationary')
```

```python
# Read the file
dataset = pd.read_csv('monthly-milk-production-pounds-p.csv')
dataset.columns=['Month', 'Milk in pounds per Cow']
dataset.drop(168, axis=0, inplace=True)

dataset['Month'] = pd.to_datetime(dataset['Month'])
dataset.set_index('Month', inplace=True)
dataset.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(dataset['Milk in pounds per Cow'], freq=12)
fig = plt.figure()
fig = decomposition.plot()
fig.set_size_inches(15,8)

from statsmodels.tsa.stattools import adfuller
#result = adfuller(dataset['Milk in pounds per Cow'])
result = adf_check(dataset['Milk in pounds per Cow'])
```

```python
dataset['Milk first difference'] = dataset['Milk in pounds per Cow'] - dataset['Milk in pounds per Cow'].shift(1)

result = adf_check(dataset['Milk first difference'].dropna())
dataset['Milk first difference'].plot()
```

```python
# ARIMA model
from statsmodels.tsa.arima_model import ARIMA
model=sm.tsa.statespace.SARIMAX(dataset['Milk in pounds per Cow'],order=(0,1,0),seasonal_order=(1,1,1,12))
results=model.fit()

dataset['forecast']=results.predict(start=150,end=168,dynamic=True)
dataset[['Milk in pounds per Cow','forecast']].plot(figsize=(12,8))

from pandas.tseries.offsets import DateOffset
future_dates = [dataset.index[-1] + DateOffset(months=x) for x in range(0,24) ]

future_dates_df = pd.DataFrame(index=future_dates[1:],columns=dataset.columns)
future_df = pd.concat([dataset,future_dates_df])

future_df['forecast'] = results.predict(start = 168, end = 188, dynamic= True)  
future_df[['Milk in pounds per Cow', 'forecast']].plot(figsize=(12, 8))
```
